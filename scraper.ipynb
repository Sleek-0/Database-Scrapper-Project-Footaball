```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36'}
page = "https://www.transfermarkt.com/al-ahed/startseite/verein/15720/saison_id/2023"
pageTree = requests.get(page, headers=headers)
pageSoup = BeautifulSoup(pageTree.content, 'html.parser')
```

```python
pageSoup
```

```python
PlayersList = []
AgeList= []
PositionsList = []
NationList=[]
ValuesList = []
```

```python
Players = pageSoup.find_all("img", {"class": "bilderrahmen-fixed lazy lazy"})
Age = pageSoup.find_all("td", {"class": "zentriert"})
p = pageSoup.find_all('table', {"class": "inline-table"})
Nationality = pageSoup.find_all("td", {"class": "zentriert"})
Values = pageSoup.find_all("td" ,{"class": "rechts hauptlink"})                                       
```

```python
Nationality
```

```python

for player in Players:
    PlayersList.append(player['alt'])


for td in Age:
    # Check if the text contains a birthdate and age in parentheses
    if '(' in td.text and ')' in td.text:
        # Extract the age using regular expression
        age_text = td.text.split('(')[1].split(')')[0]
        AgeList.append(age_text)

for player in p:
  c = player.find_all('tr')[1]
  PositionsList.append(c.find('td').text.strip())


for i in range(2, min(len(Nationality), len(PlayersList) * 3), 3):  # Ensure it stops at the length of PlayersList
    nationality_img = Nationality[i].find('img')
    if nationality_img:
        NationList.append(nationality_img['title'])
    else:
        NationList.append(None)

# Function to convert value strings to integers
def convert_value(value):
    if value == '-':
        return None
    value = value.replace('€', '').replace('k', '000').replace('m', '000000')
    return int(value)

# Extract and convert values
for value in Values:
    ValuesList.append(convert_value(value.text.strip()))

```

```python
def print_list_sizes(players_list, age_list, positions_list, nation_list, values_list):
    print(f"PlayersList size: {len(players_list)}")
    print(f"AgeList size: {len(age_list)}")
    print(f"PositionsList size: {len(positions_list)}")
    print(f"NationList size: {len(nation_list)}")
    print(f"ValuesList size: {len(values_list)}")
```

```python
print_list_sizes(PlayersList, AgeList, PositionsList, NationList, ValuesList)
```

```python
max_length = len(PlayersList)
PlayersList += [None] * (max_length - len(PlayersList))
AgeList += [None] * (max_length - len(AgeList))
PositionsList += [None] * (max_length - len(PositionsList))
NationList += [None] * (max_length - len(NationList))
ValuesList += [None] * (max_length - len(ValuesList))
```

```python
print_list_sizes(PlayersList, AgeList, PositionsList, NationList, ValuesList)
```

```python
final_df=pd.DataFrame({"Player":PlayersList,
                      "Age":AgeList,
                      "Position":PositionsList,
                      "Nation":NationList,
                      "Values":ValuesList})
```
